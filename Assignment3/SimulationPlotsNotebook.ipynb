```python
%matplotlib inline
import preferential_deletion as pp
import numpy as np

np.random.seed(12)
time_steps = 5000
t_degree = 4000
p_births = [.6, .75, .9, .8]
simulations = 5
markers = ("^", "s", "D", "D")
ylim = 4000
time_steps_collect = range(1000, 6000, 1000)

# Run Simulation
nodes_dict, edges_dict, degree_dist_dict = pp.run_simulation(p_births=p_births, simulations=simulations,
                                                          time_steps=time_steps,
                                                          time_steps_collect=time_steps_collect,
                                                          t_degree=t_degree)
```

```python
markers = ("^", "s", "D", "D")
pp.plot_expected(time_steps, nodes_dict, p_births, markers, E_func=pp.run_expected_nodes, ylim=ylim,
                  time_steps_collect=time_steps_collect, title='expected_nodes')
```

```python
pp.plot_expected(time_steps, edges_dict, p_births, markers, E_func=pp.run_expected_edges, ylim=ylim,
                  time_steps_collect=time_steps_collect, title='expected_edges')
```

```python
pp.plot_degree_dist(degree_dist_dict)
```
